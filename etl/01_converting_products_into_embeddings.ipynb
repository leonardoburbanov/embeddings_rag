````python
import base64
import cohere
import numpy as np
import urllib.request

co = cohere.ClientV2(api_key="CVjpkOY6TMQ3VIfYqBVrBiYhoninecium8PeGMxZ")


def unsplash_to_data_url(url: str) -> str:
    """Download an Unsplash image and convert it to a Cohere data URL."""
    with urllib.request.urlopen(url) as response:
        image_bytes = response.read()
    content_type = response.headers.get_content_type()
    b64 = base64.b64encode(image_bytes).decode("utf-8")
    return f"data:{content_type};base64,{b64}"


def cosine_similarity(a, b) -> float:
    """Cosine similarity between two embedding vectors."""
    a, b = np.array(a), np.array(b)
    return float(np.dot(a, b) / (np.linalg.norm(a) * np.linalg.norm(b)))
````

````python
products = [
    {
        "id": 1,
        "name": "Running Shoes Pro",
        "text": "Lightweight black running shoes with cushioned midsole",
        "unsplash_url": "https://images.unsplash.com/photo-1542291026-7eec264c27ff?w=800",
    },
    {
        "id": 2,
        "name": "Wireless Headphones",
        "text": "Over-ear headphones with active noise cancellation",
        "unsplash_url": "https://images.unsplash.com/photo-1505740420928-5e560c06d30e?w=800",
    },
    {
        "id": 3,
        "name": "Classic Leather Sneakers",
        "text": "Minimalist white leather sneakers for everyday wear",
        "unsplash_url": "https://images.unsplash.com/photo-1595950653106-6c9ebd614d3a?w=800",
    },
]
````

````python
MODEL = "embed-v4.0"

text_inputs = [f"{p['name']}. {p['text']}" for p in products]

image_inputs = [
    {
        "content": [
            {
                "type": "image_url",
                "image_url": {"url": unsplash_to_data_url(p["unsplash_url"])},
            }
        ]
    }
    for p in products
]

text_embeddings = co.embed(
    model=MODEL,
    input_type="search_document",
    texts=text_inputs,
    embedding_types=["float"],
).embeddings.float

image_embeddings = co.embed(
    model=MODEL,
    input_type="search_document",
    inputs=image_inputs,
    embedding_types=["float"],
).embeddings.float

for i, product in enumerate(products):
    product["text_embedding"] = text_embeddings[i]
    product["image_embedding"] = image_embeddings[i]
````

````python
def embed_text_query(query: str) -> list[float]:
    """Embed a text search query."""
    return co.embed(
        model=MODEL,
        input_type="search_query",
        texts=[query],
        embedding_types=["float"],
    ).embeddings.float[0]


def embed_image_query(image_url: str) -> list[float]:
    """Embed an image search query."""
    data_url = unsplash_to_data_url(image_url)
    return co.embed(
        model=MODEL,
        input_type="search_query",
        inputs=[
            {
                "content": [
                    {
                        "type": "image_url",
                        "image_url": {"url": data_url},
                    }
                ]
            }
        ],
        embedding_types=["float"],
    ).embeddings.float[0]


def rank_by_text(query: str, top_k: int = 3):
    """Search products using text only."""
    query_emb = embed_text_query(query)
    scores = [
        (p["name"], cosine_similarity(query_emb, p["text_embedding"]))
        for p in products
    ]
    return sorted(scores, key=lambda x: x[1], reverse=True)[:top_k]


def rank_by_image(image_url: str, top_k: int = 3):
    """Search products using image only."""
    query_emb = embed_image_query(image_url)
    scores = [
        (p["name"], cosine_similarity(query_emb, p["image_embedding"]))
        for p in products
    ]
    return sorted(scores, key=lambda x: x[1], reverse=True)[:top_k]


def rank_by_hybrid(
    query: str,
    image_url: str,
    text_weight: float = 0.5,
    image_weight: float = 0.5,
    top_k: int = 3,
):
    """Search products using text + image together."""
    text_query_emb = embed_text_query(query)
    image_query_emb = embed_image_query(image_url)

    scores = []
    for p in products:
        text_score = cosine_similarity(text_query_emb, p["text_embedding"])
        image_score = cosine_similarity(image_query_emb, p["image_embedding"])
        final_score = text_weight * text_score + image_weight * image_score
        scores.append((p["name"], final_score, text_score, image_score))

    return sorted(scores, key=lambda x: x[1], reverse=True)[:top_k]
````

# Resumen: tipos de vectores, búsquedas y aplicaciones

## 1. Tipos de representaciones

| Tipo | Forma | Qué captura | Almacenamiento |
|------|-------|-------------|----------------|
| **Dense (denso)** | Array de floats (`[0.12, -0.34, ...]`) | Significado semántico aprendido | Fijo (ej. 1024 × 4 bytes) |
| **Sparse (disperso)** | Pares `{índice: peso}` | Keywords + expansión lexical aprendida | Eficiente si pocos términos activos |
| **Full-text (BM25)** | Texto indexado (no es vector) | Coincidencia literal de tokens | Texto + índice invertido |
| **Cuantizado** | Dense comprimido (int8, binary) | Mismo significado, menos precisión | Menor memoria |
| **Multivector** | Varios vectores por documento/chunk | Precisión fina (ColBERT) | Alto coste |

### Gradiente de señales

```
Semántica (concepto)  ←————————————————————→  Coincidencia exacta (tokens)
     Dense                    Sparse aprendido              BM25 / Full-text
```

## 2. Tipos de búsqueda (Pinecone / RAG)

| Búsqueda | Señal | Cómo funciona | Fortaleza | Debilidad |
|----------|-------|---------------|-----------|-----------|
| **Semantic (dense)** | Embeddings | ANN + cosine/dotproduct | Sinónimos, paráfrasis, multimodal | Términos exactos, SKUs raros |
| **Full-text (BM25)** | Tokens literales | Índice invertido + BM25 | Nombres, frases, booleanos | No entiende sinónimos |
| **Sparse lexical** | Sparse aprendido | Producto interno en índices activos | Keywords inteligentes + expansión | Requiere encoder sparse |
| **Hybrid** | Dense + lexical | Combina scores (`alpha`) o filtro + ranking | Lo mejor de ambos mundos | Más complejidad, tuning |

### Filtro vs scoring

| | Filtro | Scoring |
|--|--------|---------|
| Naturaleza | Sí/No (determinista) | Ranking (probabilístico) |
| Ejemplos | `price < 50`, `$match_phrase` | BM25, cosine similarity |
| Cuándo | Antes de rankear | Ordena candidatos |

> **Nota:** precio, fecha, stock → **metadata/SQL**, no vectores.

## 3. Aplicación en búsqueda de productos

### Datos típicos por producto

```text
id, name, category, price, color, description
text_embedding   → dense (Cohere embed-v4.0)
image_embedding  → dense (opcional, multimodal)
texto crudo      → BM25 (opcional, nombres/SKUs exactos)
```

### Queries y herramienta recomendada

| Query del usuario | Herramienta | Por qué |
|-------------------|-------------|---------|
| `"zapatos cómodos para caminar"` | Dense | Intención semántica |
| `"Classic Leather Sneakers"` | BM25 / sparse | Nombre exacto |
| `"SKU-8842"` | BM25 + filtro | Identificador literal |
| Foto de zapatilla | `image_embedding` | Similitud visual |
| Texto + foto | Hybrid multimodal | Dos modalidades |
| `"los 5 más baratos"` | `ORDER BY price` | Orden numérico, no vector |
| `"electrónica < 100€"` | Filtro metadata | Condición dura |
| `"zapatillas baratas para correr"` | Dense → sort por `price` | Semántica + precio en app |

### Esquema recomendado por fase

| Fase | Qué guardar |
|------|-------------|
| **MVP** (este notebook) | 1 `text_embedding` por producto |
| **Producción básica** | Dense + texto para BM25 |
| **Multimodal** | `text_embedding` + `image_embedding` |
| **Producción completa** | Dense + BM25/sparse + metadata (`price`, `category`) |

### Lo que NO necesitas

- Varios densos del mismo texto con el mismo modelo
- Embeddings para ordenar por precio
- Chunking (los productos son documentos cortos)

## 4. Aplicación en documentos (RAG)

### Por extensión del documento

| Extensión | Ejemplo | Estrategia | Vector |
|-----------|---------|------------|--------|
| **Muy corto** | Título, FAQ, tweet | 1 doc = 1 vector | Dense o BM25 |
| **Corto–medio** | Párrafo, artículo | 1 doc = 1 vector | Dense |
| **Medio–largo** | Capítulo, manual | **Chunking** | Dense por chunk |
| **Muy largo** | PDF, libro | Chunks + metadata (`doc_id`, `page`) | Dense por chunk + hybrid opcional |

### Chunking (documentos largos)

```text
Documento largo
    ├─ chunk 1 (512 tokens) → vector₁
    ├─ chunk 2 (512 tokens) → vector₂
    └─ chunk 3 (512 tokens) → vector₃
```

| Parámetro | Valor habitual |
|-----------|----------------|
| `chunk_size` | 300–800 tokens |
| `overlap` | 10–20% |

### Queries y herramienta

| Query | Herramienta |
|-------|-------------|
| Pregunta en lenguaje natural | Dense por chunk |
| Término técnico / cláusula exacta | BM25 o sparse |
| Contrato: *"indemnización sección 4.2"* | Hybrid (dense + BM25) |
| Máxima precisión en corpus grande | Dense + reranker (o ColBERT) |

## 5. Hybrid search: patrones

| Patrón | Cuándo | Cómo combina |
|--------|--------|--------------|
| **Filtro + dense** | "Semántica pero solo docs con frase X" | `$match_phrase` + `dense_vector` |
| **Dense + sparse (mismo índice)** | Vector API, ambas señales | `alpha * dense + (1-alpha) * sparse` |
| **Dos búsquedas + merge** | Máximo control | Top-K dense + top-K BM25 → dedup → rerank |
| **Multimodal (productos)** | Texto + imagen | `rank_by_hybrid` con pesos por modalidad |

### Valores de `alpha` (dense vs sparse)

| Alpha | Sesgo | Caso de uso |
|-------|-------|-------------|
| `0.75` | Semántico | Lenguaje natural |
| `0.50` | Balanceado | Mix keyword + intención |
| `0.25` | Lexical | SKUs, IDs, nombres propios |

## 6. Árbol de decisión rápido

```text
¿Es búsqueda de productos o documentos?
│
├─ PRODUCTOS
│   ├─ ¿Query semántica? → Dense
│   ├─ ¿Nombre/SKU exacto? → BM25 / sparse
│   ├─ ¿Por imagen? → image_embedding
│   └─ ¿Por precio/stock? → Metadata (no vector)
│
└─ DOCUMENTOS
    ├─ ¿Cabe en límite del modelo? → 1 vector dense
    ├─ ¿Documento largo? → Chunking + dense por chunk
    ├─ ¿Términos exactos críticos? → + BM25/sparse por chunk
    └─ ¿Máxima precisión? → + reranker
```
